<h1>FIXING DATETIME FORMAT</h1>

We have __FlightDate__ and __CRSDepTime (Scheduled Departure Time)__, __CRSArrTime (Scheduled Arrival Time)__ columns with date and time information in our dataframe. First we will convert FlightDate to __datetime object__ and then we will create `FlightDate_Dest` because date may change when the airplane lands on the destination due to time zone difference and flight duration. Then we will create a flight DATETIME at origin (`FlightDateTime_Origin`) and destination (`FlightDateTime_Dest`)

```python
## Changing FlightDate column to datetime object
merged.FlightDate = pd.to_datetime(merged.FlightDate)
```

```python
def create_datetime_column(date_column, time_column):
    '''This function creates new datetime of flights columns at Origin and Destination'''
    
    new_col = list()
    time_column[(time_column == 2400)] = 0
    for row in range(len(date_column)):
        temp_Date = date_column[row]
        temp_Time = time_column[row]
        temp_TimeFormatted = datetime.strptime(str(temp_Time).zfill(4), "%H%M")
        temp_Hour = temp_TimeFormatted.hour
        temp_Min = temp_TimeFormatted.minute
        new_col.append(temp_Date + pd.DateOffset(hour = temp_Hour, minute = temp_Min))
    return new_col
```

```python
## Creating new FligtDateTime_Origin column from FlightDate and CRSDepTime columns
merged['FlightDateTime_Origin'] = create_datetime_column(merged.FlightDate, merged.CRSDepTime)
```

```python
## Creating a new column for the Destination the Date is being checked whether changing or not at Destination due to time zone and flight elapsed time
merged["FlightDate_Dest"] = np.where((merged.CRSArrTime - merged.CRSDepTime) > 0, 
                                    pd.DatetimeIndex(merged.FlightDate) + pd.DateOffset(0),
                                    pd.DatetimeIndex(merged.FlightDate) + pd.DateOffset(1))
```

```python
merged[["FlightDate", "Origin", "Dest", "CRSDepTime", "CRSArrTime", "FlightDate_Dest"]][merged.CRSDepTime > 2300].head(10) ## To show the date change at the destination we chose 2300 as a flight start time.
```

As seen from the above table, flight date is changing at destination.

```python
## To create FlightDateTime_Dest form FlightDate and CRSArrTime columns
merged['FlightDateTime_Dest'] = create_datetime_column(merged.FlightDate_Dest, merged.CRSArrTime)
```

```python
merged['Flight_Hour'] = merged.FlightDateTime_Origin.dt.hour
```

<h1>HANDLING WITH MISSING VALUES</h1>

```python
# Function to calculate missing values by column
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

### Infreantial Statistic

```python
fig, ax  =  plt.subplots(figsize=(12,6))
ax.plot(train_set.CRSElapsedTime, train_set.Distance, marker='o', linestyle='', ms=5, alpha=0.4)
plt.xlabel("CRSElapsedTime, mins")
plt.ylabel("Distance, miles")
_ = plt.xticks(size=13)
_ = plt.yticks(size=13)
_ = plt.xlabel('Scheduled ElapsedTime, minutes', size=14)
_ = plt.ylabel('Distance, miles ', size=14)
_ = plt.title('Elapsed Time vs Distance', size=14)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

 Visually, we can observe a strong positive correlation with few outliers. For some distances there are many CRSElapsedTime. This could be due to time-zone difference, or because of changes in reservation system for same distance flights. Here we will perform a hypothesis test to check the correlation. In this context we will start with calculating a __Pearson correlation coefficient__. But as it is known for this calculation both CRSElapsedTime and Distance have to be normally distributed. Then, let us check their distribution first. If the correlation is proved then we are supposed to include only one of both varibles into our model. 

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18,8))
sns.distplot(train_set.CRSElapsedTime, 500, ax=ax1,color='b')
ax1.set_title("CRSElapsedTime Distribution")
sns.distplot(train_set.Distance, 500, ax=ax2, color='r')
ax2.set_title("Distance Distribution")
plt.show()
```

From the figures both Scheduled Elapsed Time and Distance don't seem to be normal distributions. So we are supposed to test the p-values 

```python
from scipy import stats
stats.normaltest(train_set.CRSElapsedTime)
```

```python
stats.normaltest(train_set.Distance)
```

For both pvalue are 0 and less than 0.05 so we `can rejct the null hypothesis`. It suggests that the the distribution is not normal. __We cannot use Pearson's correlation coefficient__ in this case, but a non-parametric test such as `Spearman's test`. It is time to calculate the Spearman's correlation coefficient to check the correlation between CRSElepasedTime and Distance.

```python
stats.spearmanr(train_set.CRSElapsedTime, train_set.Distance)
```

According to the test result we can reject the null hypothesis suggesting there is no correlation between varaibales, and accept the alternate hypothesis suggesting there is pretty `strong positive correlation` (98%) between two variables. We can also confirm this correlation through heatmap.

```python
fig = plt.subplots(figsize=(12,6))
sns.set(font_scale=1.5)

corr = train_set[['CRSElapsedTime','Distance']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', annot_kws={"size": 16}, vmin=-1, vmax=1, lw=2)
_ = plt.xticks(size=16)
_ = plt.yticks(size=16)
```

After double check, now we can drop <font color='red' size=4>one of both varibles</font> from our data set.

Now it is time to find which variable is supposed to be dropped from our model. For this aim, now let's use `t-test` to check whether Cancelled varibale has an association with either CRSElapsedTime or Distance

First, we will compare the mean value of the CRSElapsedTime variable between Cancelled and Non-canceled flights. We will use  `Welch's T-test` instead of 
T-test, because as we detected earlier there is a quite __imbalance between Cancelled and Non-cancelled__.

As it is known T-test/Welch's T test requires/assumes normally distributed data. We have a problem here because as we observed earlier, the CRSElapsedTime and Distance variables are not distributed normally. In this context, first we are supossed to transform these variables into normalization. Our preference to do so is __Box-Cox transformation__.

You can see the original data distribution graphic to remember on the left and transformed data on the right graphic.

```python
_ = plt.figure()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,8))

ax1 = sns.distplot(train_set.CRSElapsedTime, 500, ax=ax1,color='b')
ax1.set_title('CRSElapsTime Distribution (Before Transform)')

ax2 = sns.distplot(stats.boxcox(train_set['CRSElapsedTime'])[0], ax=ax2, color='r')
ax2.set_title('CRSElapsedTime Distribution (After Transform)')
```

We will do the same thing for Distance variable. 

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,8))

ax1 = sns.distplot(train_set.Distance, 500, ax=ax1,color='b')
ax1.set_title('Distance Distribution (Before Transform)')

ax2 = sns.distplot(stats.boxcox(train_set['Distance'])[0], ax=ax2, color='r')
ax2.set_title('Distance Distribution (After Transform)')
```

Although both varibales do not seem perfectly normally distributed, but they look much better after Box-Cox transformation. Now, we can perform T-test with those transformed variables. 

The hypothesis of T-test for those variables:

__H0 : The means between Cancelled and Non-cancelled are the same__

__H1 : The means between Cancelled and Non-cancelled are NOT the same__

__Alpha : 0.05 ==> C.I. :0.95__

```python
t, p = stats.ttest_ind(stats.boxcox(train_set[train_set['Cancelled'] == 0].CRSElapsedTime)[0], 
                       stats.boxcox(train_set[train_set['Cancelled'] == 1].CRSElapsedTime)[0], 
                       equal_var=False)

print('CRSElapsedTime for Cancelled vs. Non-cancelled Flights')
print('------------------------------------------------------')
print('t-statistic: %s' % t)
print('p-value    : %s' % p)
```

T-test for CRSElapsedTime: The T-statistic is 963.1 and the p-value is 0. This __p-value is strictly below our alpha of 0.05__ and therefore `we can safely reject the null hypothesis`. There is statistically significant evidence that the mean CRSElapsedTime values for Cancelled and Non-canceled flights are not the same.

```python
t, p = stats.ttest_ind(stats.boxcox(train_set[train_set['Cancelled'] == 0].Distance)[0], 
                       stats.boxcox(train_set[train_set['Cancelled'] == 1].Distance)[0], 
                       equal_var=False)

print('Distance for Cancelled vs. Non-cancelled Flights')
print('------------------------------------------------------')
print('t-statistic: %s' % t)
print('p-value    : %s' % p)
```

__T-test for Distance:__ The T-statistic is 401 and the p-value is 0. This __p-value also is far below our alpha of 0.05__ and therefore `we can safely reject the null hypothesis`, as well. There is statistically significant evidence that the mean Distance values for Cancelled and Non-canceled flights are not the same.

We have already prove that __Distance and CRSElapsedTime have a strong positive correlataion__ with each other and decide to remove one of both from our model. We see that t-statistic of CRSElapsedTime is greater than t-statistic of Distance so we decided to include CRSElapseTime to our model rather than Distance

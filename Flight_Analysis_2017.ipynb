There are 302 Destination airport in our model and majority of them used seldomly during the first 6 months of 2017. In this context, only top 20 of the airports will be kept in our model.

```python
df_f.groupby('Dest')['Dest'].count().sort_values(ascending=False).head(20)
```

When we compare the top 20 aiports in terms of flights numbers, we recognized that top 20 aiports are the same both in Origin and Destinantion.

```python
## Top 20 airports have been kept in the major_airports dict

major_airports = dict(df_f.groupby('Dest')['Dest'].count().sort_values(ascending=False).head(20))
```

```python
## DataFrame is filtered according to Origin and Dest Airports. Only top 20 airports have been kept in our analysis model

df_f = df_f[df_f['Dest'].isin(major_airports.keys()) & df_f['Origin'].isin(major_airports.keys())]
```
